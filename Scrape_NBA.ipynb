```python
import requests
```

```python
standings_url = "https://www.basketball-reference.com/awards/awards_{}.html"
```

```python
years = list(range(2005, 2026))
```

```python
for year in years:
    url = standings_url.format(year)
    data = requests.get(url)

    with open("MVPS/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
tables = []
for year in years:
    with open("MVPS/{}.html".format(year)) as f:
        page = f.read()
    parser = BeautifulSoup(page, "html.parser")
    useless = parser.find('tr', class_ = "over_header")
    if useless is not None:
        useless.decompose()
    table = parser.find(id = "mvp")
    data = pd.read_html(str(table))[0]
    data["year"] = year
    tables.append(data)
```

```python
mvps = pd.concat(tables)
```

```python
mvps.head()
```

```python
mvps["Player"] = mvps["Player"].str.replace("Ä", "ć", regex = False)
mvps["Player"] = mvps["Player"].str.replace("Ä", "č", regex = False)
mvps["Player"] = mvps["Player"].str.replace("Luka Donćić", "Luka Dončić", regex = False)
mvps["Player"] = mvps["Player"].str.replace("Nikola Jokić", "Nikola Jokić", regex = False)
```

```python
import selenium
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service = service, options = options)
```

```python
import time
driver.set_page_load_timeout(120)

for year in years:
    url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)
    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(2)
    
    html = driver.page_source
    with open("All_Players/{}.html".format(year), "w+") as f:
        f.write(html)
```

```python
All_player_lst = []
for year in years:
    with open("All_Players/{}.html".format(year)) as f:
        page = f.read()
    parser = BeautifulSoup(page, "html.parser")
    parser.find('tr', class_ = "thead").decompose()
    table = parser.find(id = "switcher_per_game_stats")
    data = pd.read_html(str(table))[0]
    data["year"] = year
    All_player_lst.append(data)
```

```python
player_stats = pd.concat(All_player_lst)
```

```python
player_stats.head()
```

```python
del player_stats["Rk"]
```

```python
player_stats["Player"] = player_stats["Player"].str.replace("*", "", regex = False)
```

```python
player_stats = player_stats[player_stats["Player"] != "Player"]
```

```python
def breaker(group):
    if len(group) == 1:
        return group
    elif "2TM" in group["Team"].values:
        return group[group["Team"] == "2TM"]
    else:
        return group.iloc[[0]]
player_stats = player_stats.groupby(["Player", "year"]).apply(breaker)
```

```python
player_stats.head()
```

```python
player_stats.index = player_stats.index.droplevel()
```

```python
player_stats.index = player_stats.index.droplevel()
```

```python
player_stats = player_stats.reset_index(drop=True)
mvps = mvps.reset_index(drop=True)

Total = pd.merge(player_stats, mvps, on=["Player", "year"], how="left")
```

```python
Total = Total[Total["Team"] != "2TM"]
Total = Total[Total["Team"] != "3TM"]
Total = Total[Total["Team"] != "5TM"]
Total = Total[Total["Team"] != "4TM"]
```

```python
east = []
for year in years:
    with open("TEAMS/{}.html".format(year)) as f:
        page = f.read()
    parser = BeautifulSoup(page, "html.parser")
    parser.find('tr', class_ = "thead").decompose()
    table = parser.find(id = "divs_standings_E")
    data = pd.read_html(str(table))[0]
    data["year"] = year
    east.append(data)
west = []
for year in years:
    with open("TEAMS/{}.html".format(year)) as f:
        page = f.read()
    parser = BeautifulSoup(page, "html.parser")
    parser.find('tr', class_ = "thead").decompose()
    table = parser.find(id = "divs_standings_W")
    data = pd.read_html(str(table))[0]
    data["year"] = year
    west.append(data)
```

```python
team_stats_W = pd.concat(west)
team_stats_E = pd.concat(east)
```

```python
team_stats_W.rename(columns={"Western Conference": "Team"}, inplace=True)
team_stats_E.rename(columns={"Eastern Conference": "Team"}, inplace=True)
```

```python
team_stats = pd.concat([team_stats_W, team_stats_E])
```

```python
team_stats.head()
```

```python
team_stats = team_stats[~team_stats.Team.str.contains("Division")]
```

```python
team_stats["Team"] = team_stats["Team"].str.replace("*", "", regex = False)
```
