```python
import pandas as pd
```

```python
stats = pd.read_csv("complete_data.csv")
```

```python
stats.head()
```

```python
del stats["Unnamed: 0"]
```

```python
pd.isnull(stats).sum()
```

```python
stats = stats.fillna(0)
```

```python
stats.columns
```

```python
predictors = ['G', 'GS', 'MP', 'FG', 'FGA', 'FG%',
       '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%',
       'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS',
       'W', 'L', 'W/L%', 'PS/G',
       'PA/G', 'SRS']
```

```python
train_data = stats[stats["year"] < 2025]
test_data = stats[stats["year"] == 2025]
```

```python
x_train = train_data[predictors]
y_train = train_data["Share"]
x_test = test_data[predictors]
y_test = test_data["Share"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_mod = LinearRegression()
```

```python
lin_mod.fit(x_train, y_train)
```

```python
lin_mod.coef_
```

```python
y_pred = lin_mod.predict(x_test)
```

```python
test_data["Predicted_Share"] = y_pred
```

```python
comparer = test_data[["Player", "Share", "Predicted_Share"]]
```

```python
comparer = comparer.sort_values(by="Predicted_Share", ascending=False)
comparer["Predicted_Rank"] = list(range(1, comparer.shape[0] + 1)) 
```

```python
comparer = comparer.sort_values(by="Share", ascending=False)
comparer["Rank"] = list(range(1, comparer.shape[0] + 1)) 
```

```python
comparer.head(5)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(comparer["Share"], comparer["Predicted_Share"])
```
